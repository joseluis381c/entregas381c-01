#Solución 1
El vector de estado para este sistema cuántico compuesto por múltiples qubits se puede representar como el producto tensorial de los vectores de estado individuales de cada qubit. En este caso, tenemos tres qubits en los estados |+⟩, |+⟩ y |−⟩. Los estados |+⟩ y |−⟩ son estados de superposición de la base {|0⟩, |1⟩}, y están definidos como:

|+⟩ = (1/√2) * |0⟩ + (1/√2) * |1⟩
|−⟩ = (1/√2) * |0⟩ - (1/√2) * |1⟩

El producto tensorial de estos estados individuales nos dará el vector de estado del sistema completo:

|ψ⟩ = |+⟩ ⊗ |+⟩ ⊗ |−⟩

Calculando el producto tensorial:

|ψ⟩ = (1/√2) * |0⟩ ⊗ (1/√2) * |0⟩ ⊗ (1/√2) * |0⟩
+ (1/√2) * |0⟩ ⊗ (1/√2) * |0⟩ ⊗ -(1/√2) * |1⟩
+ (1/√2) * |1⟩ ⊗ (1/√2) * |0⟩ ⊗ (1/√2) * |0⟩
+ (1/√2) * |1⟩ ⊗ (1/√2) * |0⟩ ⊗ -(1/√2) * |1⟩

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram

# Función que define el oráculo balanceado
def balanced_oracle(qc, n):
    if n == 3:
        qc.cx(0, 3)
        qc.cx(1, 3)
        qc.cx(2, 3)
    else:
        raise ValueError("Este oráculo balanceado solo está definido para n=3")

# Crear el circuito cuántico
n = 3
qc = QuantumCircuit(n + 1, n)

# Preparación de qubits auxiliares
qc.x(n)  # Invertir el qubit de salida auxiliar para obtener el resultado correcto

# Aplicar compuertas Hadamard a los qubits de entrada
qc.h(range(n))

# Aplicar el oráculo balanceado
balanced_oracle(qc, n)

# Aplicar compuertas Hadamard a los qubits de entrada nuevamente
qc.h(range(n))

# Medición
qc.measure(range(n), range(n))

# Simulador
simulator = Aer.get_backend('qasm_simulator')
compiled_circuit = transpile(qc, simulator)
qobj = assemble(compiled_circuit, shots=1024)
result = simulator.run(qobj).result()
counts = result.get_counts()

# Mostrar el histograma de resultados
plot_histogram(counts)
```

#Solución 6:
En el contexto del algoritmo de Deutsch-Jozsa y la construcción de un oráculo balanceado, no es necesario aplicar la compuerta CX  a todos los n primeros qubits. En realidad, puedes prescindir de algunos y aún obtener un oráculo balanceado. Esto es porque la clave para que el oráculo sea balanceado radica en tener al menos un par de qubits que actúen como control y objetivo en la compuerta CX.

Dado que estás implementando una función balanceada para n = 3, solo necesitas un par de qubits para la compuerta CX. Esto se debe a que el oráculo balanceado está diseñado para tener la misma cantidad de 0 y 1 en su tabla de verdad. Por lo tanto, basta con aplicar la compuerta CX a un par de qubits, y los demás qubits no afectarán la propiedad balanceada del oráculo.

#Solución 9:
Para encontrar el estado |a⟩ tal que QFT†|a⟩ = |100⟩.


La solución sin perder generalidad, en este caso, estamos buscando un estado |a⟩ tal que QFT†|a⟩ = |100⟩. La notación |100⟩ representa un estado cuántico en el cual el primer qubit está en el estado |1⟩ y los otros dos qubits están en el estado |0⟩.

Para encontrar |a⟩, debemos aplicar la QFT inversa al estado |100⟩ y resolver para |a⟩:

QFT†|a⟩ = |100⟩

 Podemos trabajar hacia atrás a partir del estado final |100⟩ para encontrar el estado original |a⟩.

El estado |100⟩ se puede expresar en la notación de la base canónica como:

|100⟩ = |1⟩ ⊗ |0⟩ ⊗ |0⟩

Esto significa que el primer qubit está en el estado |1⟩ y los otros dos qubits están en el estado |0⟩.



#Solución 10:
Para encontrar el estado |b⟩ tal que QFT†|b⟩ = |011⟩. Dado que QFT† es la Transformada Cuántica de Fourier inversa, aplicarla a |b⟩ nos dará el estado |011⟩.

El estado |011⟩ se puede expresar como:

|011⟩ = |0⟩ ⊗ |1⟩ ⊗ |1⟩

Esto significa que el primer qubit está en el estado |0⟩ y los otros dos qubits están en el estado |1⟩.

La implementación utilizando puertas unitarias y operaciones básicas en forma general seria:

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import numpy as np

# Crear el circuito cuántico
num_qubits = 3
qft_circuit = QuantumCircuit(num_qubits)

# Aplicar las puertas adecuadas para implementar QFT†
qft_circuit.h(0)
qft_circuit.cp(np.pi/2, 0, 1)
qft_circuit.cp(np.pi/4, 0, 2)
qft_circuit.h(1)
qft_circuit.cp(np.pi/2, 1, 2)
qft_circuit.h(2)

# Simulación del circuito cuántico
simulator = Aer.get_backend('statevector_simulator')
job = assemble(transpile(qft_circuit, simulator), backend=simulator)
result = simulator.run(job).result()
statevector = result.get_statevector()

# Visualizar el resultado
plot_bloch_multivector(statevector)
```
